```python
import matplotlib.pyplot as plt
students = ["Arav", "Diya", "Kabir", "Riya", "Arjun"]
fees = [48000, 50000, 69000, 36000, 40000]
plt.barh(students, fees, color="lightpink")
plt.title("Fee paid by students")
plt.xlabel("Fee paid")
plt.ylabel("students")
plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["Arav", "Diya", "Kabir", "Riya", "Arjun"]
fees = [48000, 50000, 69000, 36000, 40000]
colors = ["lightpink", "purple", "lightblue", "lightgreen"]
plt.pie(fees, labels=students, autopct="%0.1f%%",startangle=90)
plt.title("Fee paid by students")
plt.show()
```

```python
import plotly.express as px
import pandas as pd

#sample data
data = {
    "Student": ["Arav", "Diya", "Kabir", "Riya", "Arjun"],
    "Fee Paid": [48000, 50000, 69000, 36000, 40000]
}
df = pd.DataFrame(data)

#create barchart
fig = px.bar(df, x="Student", y="Fee Paid", title="Fee Paid by Students")

#save to html file (opens directly in browser, no server)
fig.write_html("bar_chart.html", auto_open=True)
fig.show()
```
