```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


#sample data
data={
    "student": ["Arav","Diya","Kabir","Riya","Arjun"],
    "Fee paid": [48000,43200,45600,50400,48000]
}
df=pd.DataFrame(data)

#simple for chart
sns.barplot(x="student",y="Fee paid",data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

fees = [48000,43200,45600,50400,48000]

density = gaussian_kde(fees)

xs = np.linspace(min(fees),max(fees),200)

plt.plot(xs,density(xs))
plt.show()
```

```python
import matplotlib.pyplot as plt

Fees = [48000,43200,45600,50400,48000]
plt.hist(Fees)
plt.show()
```
