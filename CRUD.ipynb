# **NESTED DICTIONARY**

```python
#Create a nested dictionary with student data
student={
    "Allice":{"Math":80,"Science":90,"English":78},
    "Bob":{"Math":82,"Science":80,"English":88},
    "Charlie":{"Math":80,"Science":90,"English":96},
    "David":{"Math":96,"Science":78,"English":82},
    "Eva":{"Math":67,"Science":90,"English":79}
}
print(student)

```

```python
import pandas as pd #
data=pd.DataFrame(student)
data
```

```python
data=pd.DataFrame(student).T
data
```

```python
data["Average"]=data.mean(axis=1)
#axis=1 denotes row verse acess to the dataframe
data
```

```python
data["Math"]=data.mean(axis=0)
#axis =0 denotes column wise acess to the  dataframe
data
#The result is not a number (NaN) because there is no variable to store the average of
```

# **DATA VISUALIZATIONS**

```python
import matplotlib.pyplot as plt
#Bar chart for students average performance
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar",color="skyblue",edgecolor="black")
plt.title("Average score of Students")
plt.xlabel("Students")
plt.ylabel("Average score")
plt.show()
```

# **How to visualize subjectwise trend**

```python
plt.figure(figsize=(8,5))
for subject in["Math","Science","English"]:
    plt.plot(data.index,data[subject],marker="o",label=subject)
plt.title("Subjectwise performance")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
data[["Math","Science","English"]].plot(kind="box")
plt.title("Marks Distribution by Students")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
